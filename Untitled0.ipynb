```
df = pd.read_csv('hw3train.csv')      
df2 = pd.read_csv('hw3test.csv')
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import scipy.stats
scaler = StandardScaler()

train_x = df[['Open Price', 'High Price', 'Low Price', 'Volume']]                   # 取出訓練資料需要分析的資料欄位
train_y = df['Close Price']   # 取出訓練資料的答案
train_y=train_y.astype(int)
#train_x['Volume']=train_x['Volume']/(10**9)

test_x = df2[['Open Price', 'High Price', 'Low Price', 'Volume']]         
test_y = df2['Close Price']
test_y=test_y.astype(int)
#test_x['Volume']=test_x['Volume']/(10**9)

op=train_x['Volume'].to_numpy().reshape(-1,1)
scaler.fit(op)
train_x['Volume']= scaler.transform(op) 


op=test_x['Volume'].to_numpy().reshape(-1,1)
scaler.fit(op)
test_x['Volume']= scaler.transform(op) 

```

```
op=train_x['Open Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
train_x['Open Price']= scaler.transform(op) 


op=train_x['High Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
train_x['High Price']= scaler.transform(op) 

op=train_x['Low Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
train_x['Low Price']= scaler.transform(op) 

op=test_x['Open Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
test_x['Open Price']= scaler.transform(op) 

op=test_x['High Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
test_x['High Price']= scaler.transform(op) 

op=test_x['Low Price'].to_numpy().reshape(-1,1)
scaler.fit(op)
test_x['Low Price']= scaler.transform(op) 








#test_y=test_y*100
test_y=test_y.astype(int)
#train_y=train_y*100
train_y=train_y.astype(int)
```

```
#Logistic Regression
#https://qiita.com/0NE_shoT_/items/b702ab482466df6e5569

from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 

kf = KFold(n_splits=5,                                
           random_state=1012,
           shuffle=True)
kf.get_n_splits(train_x)

train_acc_list = []                                   
valid_acc_list = []        

dic={'Open Price':0.4, 'High Price':0.2, 'Low Price':0.2, 'Volume':0.4}

for train_index, valid_index in kf.split(train_x): 
    train_x_split = train_x.iloc[train_index]        
    train_y_split = train_y.iloc[train_index]       
    valid_x_split = train_x.iloc[valid_index]         
    valid_y_split = train_y.iloc[valid_index]         
    
    model = LogisticRegression(random_state=1012,max_iter=500,solver='saga',class_weight='balanced')
    model.fit(train_x_split, train_y_split)          
    
    train_pred_y = model.predict(train_x_split)       
    train_acc = accuracy_score(train_y_split,         
                               train_pred_y)
    valid_pred_y = model.predict(valid_x_split)       
    valid_acc = accuracy_score(valid_y_split,         
                               valid_pred_y)
    
    train_acc_list.append(train_acc)
    valid_acc_list.append(valid_acc)

print((
    'average train accuracy: {}\n' +
    'average valid accuracy: {}\n'
).format(
    np.mean(train_acc_list),                          # 輸出平均訓練準確度                           # 輸出最高訓練準確度
    np.mean(valid_acc_list)                          # 輸出平均驗證準確度                          # 輸出最高驗證準確度
))

model.fit(train_x, train_y)                     

pred_y = model.predict(test_x)                   
acc = 0      
for num in range(0,pred_y.size-1):
  if (pred_y[num+1]-pred_y[num])>0 and (test_y[num+1]-test_y[num])>0:
    acc=acc+1
  elif (pred_y[num+1]-pred_y[num])<0 and (test_y[num+1]-test_y[num])<0:
    acc=acc+1
  elif (pred_y[num+1]-pred_y[num])==0 and (test_y[num+1]-test_y[num])==0:
    acc=acc+1
acc=acc/(test_y.size-1)
acc2=accuracy_score(pred_y,test_y)
print('數字acc:{} ,股價漲跌accuracy: {}'.format(acc2,acc)) 
```
